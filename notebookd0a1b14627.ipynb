```python
import xml.etree.ElementTree as ET
import string, re
import os.path,os
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

def cleanify(root, PATH):
    '''Generates output.txt in every subdirectory of stackexchange.'''
    strings = []
    printable = set(string.printable)
    str1 = ""

    for body in root.findall('row'):
        strings.append(body.get('Body'))

    fout = open('/kaggle/working/output.txt', 'a', encoding='utf-8')
    str = ""
    for each in strings:
        each = each.lower()
        tex = each+'\n'
        str += each

    fout.write(str)
    fout.close()

dirs = os.listdir(r'/kaggle/input/resume/stack_data')
for each in dirs:
    if each == ".DS_Store":
        continue
    PATH = os.path.join('/kaggle/input/resume/stack_data', each)
    try:
        root = ET.parse(PATH).getroot()
        logging.info('Generated paragraphs for', each)
        cleanify(root, PATH)
        logging.info('SUCCESS')
    except Exception as e:
        logging.error(f'Error occurred while processing {each}: {e}')
        continue
```

```python
from bs4 import BeautifulSoup  #Crawler #Scaper
import os
import spacy
import logging
nlp = spacy.load('en_core_web_sm')
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
def get_sentences(PATH):    
    fout = open('/kaggle/working/sentence.txt', 'a', encoding='utf-8')
    with open(PATH,encoding='utf-8') as fobj:
        for line in fobj:
            if line != '\n' and line.strip().startswith('<p>'):
                try:
                    #extract tagless paragraph-text from <p> tags.
                    soup = BeautifulSoup(line.strip(),"lxml")
                    
                    #divide this tagless paragraphs into proper sentences using NLP via spacy.
                    doc = nlp(soup.p.text)
                except:
                    #If the line can't be parsed then log the line and continue to next line.
                    logging.warning(PATH,":",line,"can't be parsed.")
                    continue
                    
                #If parsed then write each sentence in the file 'sentences.txt'.
                for each in doc.sents:
                    text = each.text+'\n'
                    fout.write(each.text+'\n')
            logging.info('SUCCESS')

PATH = '/kaggle/working/output.txt'
get_sentences(PATH)
logging.info('SUCCESS')
```
